### SSH


```python
import paramiko
from getpass import getpass

#create pass prompt
thepass = getpass(prompt="Please enter SSH Password")

#host information
host = "localhost"
port = 22
#username = test 
#password = password
username = "test"
password = thepass

try:
    
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(host, port, username, password)
    
except paramiko.AuthenticationException:
    print("auth failed")
```

### Commands in SSH

```python
#run a command
#command = "hostname & ipconfig"

#Run a list of commands
command = ['netstat -an', 'hostname', 'ipconfig']

for eachCMD in command:
    #get output from the command
    stdin, stdout, stderr = ssh.exec_command(eachCMD)

    #get results from the stdout
    lines = stdout.readlines()

    #convert the list to a string
    output = ''.join(lines)
    
    print(output)
    
    #header and footer output
    sepheader = '' + '### BEGIN ' + eachCMD + ' ###\n\n'
    sepfooter = '' + '### END ' + eachCMD + ' ###\n\n'

    cmd_output = sepheader + output + sepfooter
    #print(cmd_output)
    
    with open('ssh_results.txt', 'a') as f:     
         f.write(cmd_output)


# save command to a file
    
        
        
```
